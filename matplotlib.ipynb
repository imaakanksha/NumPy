```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.plot(x,y)
plt.title("Simple Grapth")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig, axs = plt.subplots(2,3)
axs[0,0].plot(x,y)
axs[1,1].scatter(x,y)
plt.show()
```

```python
plt.plot(x,y, linestyle='--', color='r', marker='o', label='Line')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y, color='blue')
plt.show()
```

```python
categories = ['A','B','C','D']
values = [10,20,30,40]
plt.bar(categories, values, color='orange')
plt.show()
```

```python
data=[1,1,1,2,2,2,3,4,5,6,7,8]
plt.hist(data, bins=5, color="green")
plt.show()
```

```python
sizes=[25,35,20,20]
labels = ['Offices', 'Houses', 'Malls', 'Resturants']
plt.pie(sizes, labels = labels, autopct='%1.1f%%')
plt.show()
```

```python
data=[7,8,8,9,10,10,11,11,11,11,11,12]
plt.boxplot(data)
plt.show()
```
