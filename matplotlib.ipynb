```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.plot(x,y)
plt.title("Simple Grapth")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig, axs = plt.subplots(2,3)
axs[0,0].plot(x,y)
axs[1,1].scatter(x,y)
plt.show()
```

```python
plt.plot(x,y, linestyle='--', color='r', marker='o', label='Line')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y, color='blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.grid(True)
plt.annotate("Peak", xy=(3,25), xytext=(2,30), arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()
```

```python
#Subplots
fig, axs = plt.subplots(2)
axs[0].plot(x,y)
axs[1].scatter(x,y)
plt.show()
```

```python
##Logrithmic Scales
plt.plot(x,y)
plt.yscale('log')
plt.show()
```

```python
#Error Bars
errors=[1]
plt.errorbar(x,y,yerr=errors,fmt='o', ecolor='red')
plt.show()
```

```python
#3D Visualizations

from  mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax=fig.add_subplot(111, projection='3d')
x=[1,2,3,4]
y=[10,20,25,35]
z=[5,15,25,35]
ax.scatter(x,y,z)
plt.show()
```

```python
import pandas as pd

data = {'Year' : [2000, 2005, 2010, 2015, 2020], 
        'Population' :[2.5,2.8,3.2,3.6,4.0]}
df = pd.DataFrame(data)
plt.plot(df['Year'], df['Population'])
plt.title("Population Growing Over Years")
plt.show()
        
```

```python
#Financial Data
stock_prices = [100,105,110,120,125]
days =[1,2,3,4,5]
plt.plot(days, stock_prices)
plt.title("Stock Price Over Time")
plt.xlabel("Days")
plt.ylabel("Stock Price")
plt.show()
```

```python
#Scientific Data Representation

import numpy as np
x = np.linspace(0,10.100)
y = np.sin(x)
plt.plot(x,y)
plt.title("Sine Wave")
plt.show()
```

```python
# Machine Learning Visualization
from sklearn.datasets import make_blobs
data, labels = make_blobs(n_samples=100, centers=3, random_state=42)
plt.scatter(data[:,0], data[:,1],c=labels, cmap='viridis')
plt.title("Cluster VisualiZation")
plt.show()
```
